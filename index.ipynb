## Step 4: Run a simple regression in Statsmodels with TV as a predictor

```python
# import libraries
#I did it above but I'll put it here as well.
import statsmodels.api as sm
from statsmodels.formula.api import ols
plt.style.use('seaborn')
# build the formula 
f = 'sales~TV'
# create a fitted model in one line
model = ols(formula=f, data=df).fit()
```

## Step 5: Get Regression Diagnostics Summary

```python
model.summary()
```

Note here that the coefficients represent associations, not causations

## Step 6:  Draw a prediction line with data points on a scatter plot for X (TV) and Y (Sales)

Hint: You can use the `model.predict()` function to predict the start and end point of of regression line for the minimum and maximum values in the 'TV' variable. 

```python
# create a DataFrame with the minimum and maximum values of TV
X_tv = pd.DataFrame({'TV': [df.TV.min(), df.TV.max()]})
print(X_tv.head())

# make predictions for those x values and store them
prediction = model.predict(X_tv)
print(prediction)
# first, plot the observed data and the least squares line
df.plot(kind='scatter', x='TV', y='sales')
plt.plot(X_tv, prediction, c='green', linewidth=2)
plt.show()
```

## Step 7: Visualize the error term for variance and heteroscedasticity

```python
fig = plt.figure(figsize=(15,8))
fig = sm.graphics.plot_regress_exog(model, 'TV', fig=fig)
plt.show()
```

```python
# Record Your observations on heteroscedasticity
#We can see that the line fits the data fairly well.  
#The residual plot we can see that it is the distances are well
#distributed.  Since there isn't a pattern the homoskedasticity holds
```

## Step 8: Check the normality assumptions by creating a QQ-plot

```python
# Code for QQ-plot here
import scipy.stats as stats
residuals = model.resid
fig = sm.graphics.qqplot(residuals, dist=stats.norm, line='45', fit=True)
plt.show()
```

```python
# Record Your observations on the normality assumption
#The points are almostall on the line meaning we can assume normalcy
```

## Step 9: Repeat the above for radio and record your observations

```python
# code for model, prediction line plot, heteroscedasticity check and QQ normality check here
# build the formula 
f = 'sales~radio'
# create a fitted model in one line
model = ols(formula=f, data=df).fit()
```

```python
model.summary()
```

```python
# create a DataFrame with the minimum and maximum values of TV
X_radio = pd.DataFrame({'radio': [df.radio.min(), df.radio.max()]})
print(X_radio.head())

# make predictions for those x values and store them
prediction = model.predict(X_radio)
print(prediction)
# first, plot the observed data and the least squares line
df.plot(kind='scatter', x='radio', y='sales')
plt.plot(X_radio, prediction, c='green', linewidth=2)
plt.show()
```

```python
fig = plt.figure(figsize=(15,8))
fig = sm.graphics.plot_regress_exog(model, 'radio', fig=fig)
plt.show()
```

```python
residuals = model.resid
fig = sm.graphics.qqplot(residuals, dist=stats.norm, line='45', fit=True)
plt.show()
```

```python
# Record your observations here for goodnes of fit 
#We can see that compared to TV, radio isn't as linearly
#correlated, but it is relatively close.  Based on the residuals
#as well as the QQ plot we can see linear is an ok fit.
#The Rsquared value though tells us a different story than our 
#eyes.  The Rsquared value is .33 which is very low meaning that
#the line accounts for 33% of the data.  
```
